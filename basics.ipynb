```python
new_obj=wave.open("new_audio.wav","wb")
new_obj.setnchannels(1)
new_obj.setsampwidth(2)
new_obj.setframerate(44100)
new_obj.writeframes(frames)
new_obj.close()
```

**Plotting audio signals**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
obj=wave.open("new_audio.wav", "rb")

frames=obj.readframes(-1)

time=obj.getnframes() / obj.getframerate()
print(time)
print(obj.getnframes())

x=np.linspace(0,time,num=obj.getnframes())
y=np.frombuffer(frames, dtype=np.int16)


plt.plot(x,y)
```

**Convert wav file to mp3 file**

```python
from pydub import AudioSegment
```

```python
audio= AudioSegment.from_wav("new_audio.wav")

audio=audio + 30
audio.export("audio_high_vol.mp3", format="mp3")
```
